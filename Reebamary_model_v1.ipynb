```python
import numpy as np
from sklearn import neighbors
from sklearn.model_selection import train_test_split
import pandas as pd
```

```python
#data cleaning
Breast_cancer=pd.read_csv('Breast_cancerdata.csv')
Breast_cancer.drop(['id'], 1, inplace=True)
Breast_cancer.drop(['Unnamed: 32'], 1, inplace=True)
Breast_cancer.replace(np.nan, 0)
Breast_cancer.replace(np.NaN, 0)
Breast_cancer.head()
```

```python
#Define x and y variable
x = Breast_cancer.drop('diagnosis',axis=1).to_numpy()
y = Breast_cancer['diagnosis'].to_numpy()

# Create Train and Test Datasets
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.20,stratify=y,random_state=100)

print(x_train)
```

```python
#Model building using KNN
clf = neighbors.KNeighborsClassifier()
clf.fit(x_train, y_train)

# test accuracy of the classifier
accuracy = clf.score(x_test, y_test)
print('Accuracy:', accuracy)
```
